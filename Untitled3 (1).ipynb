```python
import os
os.chdir('/content/BAMT/')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from bayesian.train_bn import structure_learning, parameter_learning
from preprocess.discretization import get_nodes_type, discretization, code_categories, get_nodes_sign
from bayesian.save_bn import save_structure, save_params, read_structure, read_params
from bayesian.sampling import generate_synthetics
from bayesian.calculate_accuracy import calculate_acc
from external.libpgm.hybayesiannetwork import HyBayesianNetwork
from visualization.visualization import draw_BN
```

```python
data = pd.read_csv('/bank.csv',delimiter=';')
data
```

# Новый раздел

```python
columns = ['age', 'job', 'marital','education', 'balance', 'day','duration', 'month', 'previous', 'pdays', 'campaign']
data = data[columns]
data.dropna(inplace=True)
data.reset_index(inplace=True, drop=True)
data['age'] = data['age'].apply(lambda x: float(x))
data['balance'] = data['balance'].apply(lambda x: float(x))
data['day'] = data['day'].apply(lambda x: float(x))
data['duration'] = data['duration'].apply(lambda x: str(x))
data['previous'] = data['previous'].apply(lambda x: str(x))
data['pdays'] = data['pdays'].apply(lambda x: str(x))
data['campaign'] = data['campaign'].apply(lambda x: str(x))
data.dropna(inplace=True)
data.reset_index(inplace=True, drop=True)
data
```

```python
nodes_type = get_nodes_type(data)
nodes_type
```

```python
nodes_sign = get_nodes_sign(data)
nodes_sign
```

```python
coded_data, coder = code_categories(data, 'label', ['campaign','pdays', 'previous', 'month', 'duration', 'education', 'marital', 'job'])
discrete_data, coder = discretization(coded_data, 'equal_frequency', ['age','balance', 'day'])
```

```python
bn_HC = structure_learning(discrete_data, 'HC', nodes_type, 'K2')
bn_HC
```

```python
draw_BN(bn_HC, nodes_type, 'houses_bn_HC')
```

```python
params_HC = parameter_learning(data, nodes_type, bn_HC, 'simple')
save_structure(bn_HC, 'houses_net_structure_HC')
skel_HC = read_structure('houses_net_structure_HC')
save_params(params_HC, 'houses_net_param_HC')
params_HC = read_params('houses_net_param_HC')
houses_bn_HC = HyBayesianNetwork(skel_HC, params_HC)
synth_df_HC = generate_synthetics(houses_bn_HC, nodes_sign, 'simple', 1200)

sns.set(style      = 'whitegrid',
        font_scale = 1.2,
        rc         = {'figure.figsize':(30, 10)})

fix, ax = plt.subplots(3, 1, figsize=(10, 12))

sns.distplot(data['balance'], ax=ax[0])
sns.distplot(synth_df_HC['balance'], ax=ax[0])
ax[0].legend(['Real data', 'Synthesized data'])

sns.distplot(data['duration'], ax=ax[1])
sns.distplot(synth_df_HC['duration'], ax=ax[1])
ax[1].legend(['Real data', 'Synthesized data'])

sns.distplot(data['age'], ax=ax[2])
sns.distplot(synth_df_HC['age'], ax=ax[2])
ax[2].legend(['Real data', 'Synthesized data'])

print(calculate_acc(houses_bn_HC, data, ['balance', 'duration', 'age'], 'simple')[0:2])
```
